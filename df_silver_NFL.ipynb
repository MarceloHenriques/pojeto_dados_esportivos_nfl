```python
import pandas as pd
import ast
```

```python
# Carregar os dados do CSV
df_silver = pd.read_csv('data_silver.csv')

# Exibir as primeiras linhas do DataFrame para verificar a estrutura
df_silver.head()
```

```python
df_silver.info()
```

```python
df_silver['team'] = df_silver['team'].apply(lambda x: ast.literal_eval(x)['name'] if isinstance(x, str) else x)
df_silver['league'] = df_silver['league'].apply(lambda x: ast.literal_eval(x)['name'] if isinstance(x, str) else x)
```

```python
df_silver.head()
```

```python
df_silver = df_silver.drop(['conference', 'records', 'streak', 'ncaa_conference'],axis=1)
df_silver
```

```python
df_silver = df_silver.drop(['ties', 'records', 'streak', 'ncaa_conference'],axis=1)
```

```python
# Separar a coluna 'points' em três novas colunas
df_silver['points_for'] = df_silver['points'].apply(lambda x: eval(x)['for'])
df_silver['points_against'] = df_silver['points'].apply(lambda x: eval(x)['against'])
df_silver['points_difference'] = df_silver['points'].apply(lambda x: eval(x)['difference'])

# Remover a coluna original 'points'
df_silver = df_silver.drop(columns=['points'])

# Exibir o DataFrame atualizado
df_silver.head()


```

```python
df_silver = df_silver.drop(columns=['ties'])
df_silver
```

```python
# Salvar o DataFrame atualizado em um novo arquivo CSV
df_silver.to_csv("df_gold_nfl.csv", index=False)
```
